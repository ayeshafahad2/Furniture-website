```python
import random
import json
import time

# File to store the leaderboard
LEADERBOARD_FILE = "leaderboard.json"

def load_leaderboard():
    """Load leaderboard from file."""
    try:
        with open(LEADERBOARD_FILE, "r") as file:
            return json.load(file)
    except (FileNotFoundError, json.JSONDecodeError):
        return {}

def save_leaderboard(leaderboard):
    """Save leaderboard to file."""
    with open(LEADERBOARD_FILE, "w") as file:
        json.dump(leaderboard, file, indent=4)

def update_leaderboard(player_name, attempts):
    """Update leaderboard with the best score."""
    leaderboard = load_leaderboard()

    if player_name not in leaderboard or attempts < leaderboard[player_name]:
        leaderboard[player_name] = attempts
        print(f"🎖 New best score: {attempts} attempts!")

    save_leaderboard(leaderboard)

def display_leaderboard():
    """Display leaderboard."""
    leaderboard = load_leaderboard()
    print("\n🏆 LEADERBOARD 🏆")
    if not leaderboard:
        print("No scores recorded yet.")
    else:
        for player, score in leaderboard.items():
            print(f"👤 {player}: {score} attempts")

def get_hint(guess, target):
    """Provide hints based on how close the guess is to the target number."""
    difference = abs(target - guess)

    if difference == 0:
        return "🎯 Spot on!"
    elif difference <= 3:
        return "🔥 Hot!"
    elif difference <= 7:
        return "🌡 Warm!"
    else:
        return "❄ Cold!"

def play_game():
    print("\n🎮 Welcome to the Number Guessing Game!")
    player_name = input("Enter your name: ").strip()

    while True:
        print("\nChoose difficulty level:")
        print("1️⃣ Easy (7 attempts)")
        print("2️⃣ Medium (5 attempts)")
        print("3️⃣ Hard (3 attempts)")
        print("📊 4️⃣ View Leaderboard")

        choice = input("Enter your choice (1/2/3/4): ")

        levels = {"1": (7), "2": (5), "3": (3)}

        if choice == "4":
            display_leaderboard()
            continue

        if choice not in levels:
            print("⚠ Invalid choice! Select 1, 2, 3, or 4.")
            continue

        attempts = levels[choice]
        number_to_guess = random.randint(50, 100)
        print(f"\n🎯 The computer has chosen a number! You have {attempts} attempts to guess it.")

        guess_counter = 0

        while guess_counter < attempts:
            try:
                user_guess = int(input("\nEnter your guess (between 50-100): "))

                if user_guess < 50 or user_guess > 100:
                    print("⚠ Please guess a number between 50 and 100!")
                    continue

                guess_counter += 1
                hint = get_hint(user_guess, number_to_guess)

                if user_guess == number_to_guess:
                    print(f"🎉 Congratulations {player_name}! You guessed the number {number_to_guess} in {guess_counter} attempts! 🏆")
                    update_leaderboard(player_name, guess_counter)
                    break
                elif guess_counter == attempts:
                    print(f"😞 You've used all attempts. The correct number was {number_to_guess}. Better luck next time!")
                else:
                    print(f"{hint} Try again! ({attempts - guess_counter} attempts left)")

            except ValueError:
                print("🚨 Invalid input! Please enter a valid number.")

        play_again = input("\n🔄 Do you want to play again? (yes/no): ").strip().lower()
        if play_again != "yes":
            print("👋 Thanks for playing! See you next time!")
            break

play_game()
```
